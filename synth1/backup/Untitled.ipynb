```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
scores1 = pd.read_excel('./BBF 1 Scoring.xlsx')
scores1['Score'] = scores1['Score'].astype("category", categories=["Standard", "Difficult", "Impossible"], ordered=True)
sns.countplot(scores1['Score'])
plt.xlabel("Synthesis difficulty")
plt.ylabel("Number of fragments")
sns.despine()

plt.savefig('./Scores1.svg')
```

```python
scores1['Score'].value_counts()
```

```python
sns.set(style="white", context="poster", font="Gill Sans Std", rc={ 'font.weight': 'light' })

scores = pd.read_csv('./Q-08034_BBF Synth1_ScoreResults.csv')
sns.countplot(scores['difficulty_label'])
plt.xlabel("Synthesis difficulty")
plt.ylabel("Number of fragments")
sns.despine()

plt.savefig('./Scores.svg')
```
